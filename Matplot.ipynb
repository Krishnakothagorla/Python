```python
##Draw a line in a diagram from position (0,0) to position (10,25):
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 10])
ypoints = np.array([0, 25])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
##Draw a line in a diagram from position (1, 3) to position (12, 15):
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 12])
ypoints = np.array([3, 15])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```
